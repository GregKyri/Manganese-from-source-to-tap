# ***Treatment technologies***

```python
def treattech(treatstage, treatmentprocess, treatalt=None):

    list_options = []

    if treatstage == "Screens":
          treatmenttechnologies = "Screens"

    elif treatstage == "Trickling Filtration":
          treatmenttechnologies= "Trickling Filtration"

    elif treatstage == "Microsieves":
          treatmenttechnologies= "Microsieves"

    elif treatstage == "Infiltration":
          treatmenttechnologies= "Infiltration"

    elif treatstage == "Coagulation":
          list_options = ["Cascade Mixing", "Inline Mixing", "Dynamic Mixing"]
          print("Coagulation:")
          print("1. Cascade Mixing")
          print("2. Inline Mixing")
          print("3. Dynamic Mixing")
          choice = int(input("Enter your choice (1/3): "))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]
          else:
              treatmenttechnologies = list_options[2]

    elif treatstage == "Flocculation":
          list_options = ["Static mixing", "Dynamic mixing"]
          print("Flocculation:")
          print("1. Static mixing")
          print("2. Dynamic Mixing")
          choice = int(input("Enter your choice (1/2): "))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]

    elif treatstage == "Floc Removal":
          list_options = ["Sludge Blanket Clarification", "Horizontal Settling", "Titled-plate Settling", "Dissolved Air Flotation"]
          print("Floc Removal:")
          print("1. Sludge Blanket Clarification")
          print("2. Horizontal Settling")
          print("3. Titled-plate Settling")
          print("4. Dissolved Air Flotation")
          choice = int(input("Enter your choice (1/4): "))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]
          elif choice==3:
              treatmenttechnologies = list_options[2]
          else:
              treatmenttechnologies = list_options[3]

    elif treatstage == "Sand Filtration":
          list_options = ["Single Media Filtration", "Dual Media Filtration","Moving Bed Filtration"]
          print("Sand Filtration:")
          print("1. Single Media Filtration")
          print("2. Dual Media Filtration")
          print("3. Moving Bed Filtration")
          choice = int(input("Enter your choice (1/3):"))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]
          else:
              treatmenttechnologies = list_options[2]

    elif treatstage == "OMP Removal":
          if treatalt == 0:
              list_options = ["Granular Activated Carbon Filtration (GAC)", "Powdered Activated Carbon Dosing (PAC)", "MP-UV/H2O2", "NF/RO", "Ozonation"]
              print("OMP Removal:")
              print("1. Granular Activated Carbon Filtration (GAC)")
              print("2. Powdered Activated Carbon Dosing (PAC)")
              print("3. MP-UV/H2O2")
              print("4. NF/RO")
              print("5. Ozonation")
              choice = int(input("Enter your choice (1/5):"))
              if choice==1:
                  treatmenttechnologies = list_options[0]
              elif choice==2:
                  treatmenttechnologies = list_options[1]
              elif choice==3:
                  treatmenttechnologies = list_options[2]
              elif choice==4:
                  treatmenttechnologies = list_options[3]
              else:
                  treatmenttechnologies = list_options[4]
              if treatmenttechnologies=="Ozonation":
                 treatmentprocess.remove("Disinfection")
              else:
                 treatmentprocess==treatmentprocess

          else:
              list_options = ["Granular Activated Carbon Filtration (GAC)", "Powdered Activated Carbon Dosing (PAC)", "NF/RO"]
              print("OMP Removal:")
              print("1. Granular Activated Carbon Filtration (GAC)")
              print("2. Powdered Activated Carbon Dosing (PAC)")
              print("3. NF/RO")
              choice = int(input("Enter your choice (1/3):"))
              if choice==1:
                  treatmenttechnologies = list_options[0]
              elif choice==2:
                  treatmenttechnologies = list_options[1]
              else:
                  treatmenttechnologies = list_options[2]

    elif treatstage == "NOM removal":
          list_options = ["Anionic IEX", "Enhanced Coagulation"]
          print("NOM removal:")
          print("1. Anionic IEX")
          print("2. Enhanced Coagulation")
          choice = int(input("Enter your choice (1/2):"))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]

    elif treatstage == "Softening":
          list_options = ["Cationic IEX", "Pellet Softening", "NF/RO"]
          print("Softening:")
          print("1. Cationic IEX")
          print("2. Pellet Softening")
          print("3. NF/RO")
          choice = int(input("Enter your choice (1/3):"))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]
          else:
              treatmenttechnologies = list_options[2]

    elif treatstage == "Aeriation":
          list_options = ["Cascades", "Spray Aeriation", "Stripping Tower", "Plate Aeration"]
          print("Aeriation:")
          print("1. Cascades")
          print("2. Spray Aeriation")
          print("3. Stripping tower")
          print("4. Plate aeration")
          choice = int(input("Enter your choice (1/4):"))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]
          elif choice==3:
              treatmenttechnologies = list_options[2]
          else:
              treatmenttechnologies = list_options[2]

    elif treatstage == "Conditioning":
          list_options = ["Remineralisation", "Dosing of Chemicals", "pH Correction"]
          print("Conditioning:")
          print("1. Remineralisation")
          print("2. Dosing of Chemicals")
          print("3. pH Correction")
          choice = int(input("Enter your choice (1/3):"))
          if choice==1:
              treatmenttechnologies = list_options[0]
          elif choice==2:
              treatmenttechnologies = list_options[1]
          else:
              treatmenttechnologies = list_options[2]

    elif treatstage == "Disinfection":
          list_options = ["Ozonation", "LP-UV", "Chlorination", "Chloramination"]
          print("Disinfection:")
          print("1. Ozonation")
          print("2. LP-UV")
          print("3. Chlorination")
          print("4. Chloramination")
          choice = int(input("Enter your choice (1/4):"))
          if choice==1:
             treatmenttechnologies = list_options[0]
          elif choice==2:
             treatmenttechnologies = list_options[1]
          elif choice==3:
             treatmenttechnologies = list_options[2]
          elif choice==4:
             treatmenttechnologies = list_options[3]
          else:
             treatmenttechnologies = list_options[4]

    elif treatstage == "Ultrafiltration":
          list_options = ["UF", "MF"]
          highNOM = input("NOM concentration in the water (mg/l): ")
          highNOM = float(highNOM)
          if highNOM > 9.8 :
             treatmenttechnologies = list_options[0]
          else:
             treatmenttechnologies = list_options[1]
    else:
          print("Error", "Invalid treatstage value")
          return []

    return treatmenttechnologies
```
